```python
def Hand (img):
  import cv2
  import matplotlib.pyplot as plt
  from google.colab.patches import cv2_imshow
  import mediapipe as mp

  mp_hands = mp.solutions.hands

  hands = mp_hands.Hands()


  hand_draw = mp.solutions.drawing_utils

  drawSpec = hand_draw.DrawingSpec(thickness=1, circle_radius= 2 )
  fr =cv2.imread(img)
 

  rgb_frame = cv2.cvtColor(fr, cv2.COLOR_BGR2RGB)
    
  result = hands.process(rgb_frame)
##
  if result.multi_hand_landmarks is not None:
       
        for hand in result.multi_hand_landmarks:

            hand_draw.draw_landmarks(fr,hand,mp_hands.HAND_CONNECTIONS,drawSpec,drawSpec)
            
  return cv2_imshow(fr)  

    


 
```
